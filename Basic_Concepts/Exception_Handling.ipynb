```python
blog_posts = [{'Photos': 3, 'Likes': 21, 'Comments': 2}, {'Likes': 13, 'Comments': 2, 'Shares': 1}, {'Photos': 5, 'Likes': 33, 'Comments': 8, 'Shares': 3}, {'Comments': 4, 'Shares': 2}, {'Photos': 8, 'Comments': 1, 'Shares': 1}, {'Photos': 3, 'Likes': 19, 'Comments': 3}]
total_likes = 0
for post in blog_posts:
  try:
   total_likes += post['Likes']
  except KeyError:
    total_likes += 0
print(total_likes)
```

```python
a1 = list(map(int,input("Enter list1 items one by one : ").split()))
a2 = list(map(int,input("Enter list2 items one by one : ").split()))
a3 = []
m = max(len(a1),len(a2))
for i in range(m):
  try:
    a3.append(a1[i]+a2[i])
  except IndexError:
    if(i>=len(a1)):
      a3.append(a2[i])
    elif(i>=len(a2)):
      a3.append(a1[i])
print(a3)
```
