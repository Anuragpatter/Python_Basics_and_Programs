```python
#18
import csv
file=open("STD record.csv")
csvfile=csv.reader(file)
data={}
for row in csvfile:
  data[row[0]]=[row[0],row[1],row[6],(row[5].split(sep=',')[0]),(row[5].split(sep=',')[1]),(row[5].split(sep=',')[2])]
prev=-1
cur=-1
m=0 
for row in data:
  d=data[row]
  prev=cur
  cur=(float(d[3])+float(d[4])+float(d[5]))/3
  x=m
  if(x<max(prev,cur)):
    m=max(prev,cur)
  print("Highest Average marks in Physics, Chemistry, Maths:","{:.2f}".format(m),"\nhighest scorer list :\nenroll    name")
  for row in data:
    d=data[row]
    if(m==(float(d[3])+float(d[4])+float(d[5]))/3):
      print(d[0]," ",d[1])
    file.close()
```

```python
#22
prime_sum = {}
def digit_sum(n):
  s = 0
  while n > 0:
    s = s+(n % 10)
    n = n//10
  return s
for x in range(50, 101):
  c = True
  for i in range(2, round(x**(0.5)+1)):
    if x % i == 0:
      c = False
      break
  if c:
    ps = digit_sum(x)
    prime_sum.update([(x, ps)])
print(prime_sum)
```
